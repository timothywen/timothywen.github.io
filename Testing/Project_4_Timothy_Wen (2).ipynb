**Impact of Clothing Color on Severity of Crash**

Description of dataset:

In this part, I analyze the rates of dark and light clothing for crash reports that result in serious injuries or fatal injuries for pedestrians. My belief is that a person wearing darker colors will have higher rates of severe crashes than pedestrians in light clothing because darker colors tend to blend in with the surrounding objects/landscape. Vivid colors such as yellow or orange may be more noticeable than black or brown.

```python

map_osm = folium.Map(location=[39.15, -77.14], zoom_start=11)

for index, crash in crash_table[crash_table['Pedestrian Type'] == 'PEDESTRIAN'].iterrows():
    if crash['Injury Severity'] == 'SUSPECTED SERIOUS INJURY' or crash['Injury Severity'] == 'FATAL INJURY':
      if crash['Pedestrian Visibility'] == 'DARK CLOTHING':
        folium.Marker(location=[crash["Latitude"], crash["Longitude"]],icon=folium.Icon(color='black', icon="empty")).add_to(map_osm)
      if crash['Pedestrian Visibility'] == 'LIGHT CLOTHING':
        folium.Marker(location=[crash["Latitude"], crash["Longitude"]],icon=folium.Icon(color='white', icon="user")).add_to(map_osm)


map_osm
```

Description of Interactive Map:

Through a preliminary analysis of the map, we can observe that there does appear to be a greater amount of severe crashes involving pedestrians wearing darker clothes compared to those wearing lighter colors. Additionally, an unintended observation that can be made is that a majority of the severe/deadly crashes are located along highways, busy intersections/roads, and/or routes. One good example of this would be along Georgia Ave. near Wheaton where we can observe a cluster of non-motorist crashes along several major freeways.

This data could be effectively utilized by officials to increase visibility and lighting in clustered areas. Additionally, another solution could be painting colors that would contrast with darker clothes to help drivers be more careful.
